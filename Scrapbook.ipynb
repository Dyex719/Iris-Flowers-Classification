```python
x1 = np.array(iris_df["SepalLengthCm"]).reshape(150,1)
x2 = np.array(iris_df["PetalLengthCm"]).reshape(150,1)
# To join vertically
X = np.concatenate((x1,x2),axis=1)
y = np.array(iris_df["Species"])
svc = svm.SVC(kernel='linear').fit(X, y)
```
